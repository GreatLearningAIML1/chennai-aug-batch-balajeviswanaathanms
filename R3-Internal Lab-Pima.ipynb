# Q8 Split the pima_df into training and test set in the ratio of 70:30 (Training:Test).</b>


```python
from sklearn.model_selection import train_test_split
array = pima_df.values
X = array[:,0:7] # select all rows and first 8 columns which are the attributes   X rep Independent
Y = array[:,8]   # select all rows and the 8th column which is the classification "Yes", "No" for diabeties Y rep traget var
test_size = 0.30 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

# Q9 Create the decision tree model using “entropy” method of reducing the entropy and fit it to training data.</b>



```python
dt_model = DecisionTreeClassifier(criterion = 'entropy' )
dt_model.fit(X_train, y_train)
```

# Q10 Print the accuracy of the model & print the confusion matrix 

```python
from sklearn import metrics
y_predict = dt_model.predict(X_test)
print(metrics.confusion_matrix(y_test, y_predict))
```

# Q12 Apply the Random forest model and print the accuracy of Random forest Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators = 60)
rfcl = rfcl.fit(X_train, y_train)
```

```python
y_predict = rfcl.predict(X_test)
print(rfcl.score(X_test , y_test))
print(metrics.confusion_matrix(y_test, y_predict))
```
