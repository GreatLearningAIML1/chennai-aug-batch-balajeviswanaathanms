### Consider only first 1000 rows in the dataset for building feature set and target set

```python
data1=data.iloc[0:1000,:]
```

### Divide the data into train and test sets

```python
# define the fetures and dependent variable with full  model
y = data1[['close']]
X = data1.drop('close', axis=1)
```

```python
# Split X and y into training and test set in 70:30 ratio
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
x_train.shape
```

```python
y_train.shape
```

## Building the graph in tensorflow

1.Define input data placeholders

```python
#Input features
x = tf.placeholder(shape=[None,4],dtype=tf.float32, name='x-input')

x_n = tf.layers.batch_normalization(x, training=True)

#Actual Prices
y_ = tf.placeholder(shape=[None,1],dtype=tf.float32, name='y-input')
```

2.Define Weights and Bias

```python
W = tf.Variable(tf.zeros(shape=[4,1]), name="Weights")
b = tf.Variable(tf.zeros(shape=[1]), name="Bias")
```

3.Prediction

```python
y = tf.add(tf.matmul(x_n,W),b,name='output')
```

4.Loss (Cost) Function [Mean square error]

```python
loss = tf.reduce_mean(tf.square(y-y_),name='Loss')
```

5.GradientDescent Optimizer to minimize Loss [GradientDescentOptimizer]

```python
train_op = tf.train.GradientDescentOptimizer(0.01).minimize(loss)
```

## Execute the Graph for 100 epochs and observe the loss

```python
#Lets start graph Execution
sess = tf.Session()

# variables need to be initialized before we can use them
sess.run(tf.global_variables_initializer())

#how many times data need to be shown to model
training_epochs = 100
```

```python
for epoch in range(training_epochs):
            
    #Calculate train_op and loss
    _, train_loss = sess.run([train_op, loss],feed_dict={x:x_train, 
                                                        y_:y_train})
    
    if epoch % 10 == 0:
        print ('Training loss at step: ', epoch, ' is ', train_loss)
```

### Load the given Iris data using pandas (Iris.csv)

```python
Iris = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/Iris.csv')
```

```python
Iris.info()
```

```python
Iris.drop(columns=["Id"],axis=1,inplace=True)
```

```python
target=iris["Species"]
feature=iris.drop(columns="Species",axis=1)
```

### Splitting the data into feature set and target set

### Target set has different categories. So, Label encode them. And convert into one-hot vectors using get_dummies in pandas.

```python
target=pd.get_dummies(target)
```

### Divide the dataset into Training and test (70:30)

```python
X_train,X_test,y_train,y_test=train_test_split(feature, target, test_size=0.30, random_state=42)
```

### Model
Build the model with following layers: <br>
1. First dense layer with 10 neurons with input shape 4 (according to the feature set) <br>
2. Second Dense layer with 8 neurons <br>
3. Output layer with 3 neurons with softmax activation (output layer, 3 neurons as we have 3 classes) <br>
4. Use SGD and categorical_crossentropy loss 

```python
#Initialize Sequential model
model = tf.keras.models.Sequential()

#Reshape data from 2D to 1D -> 28x28 to 784
#model.add(tf.keras.layers.Reshape((420,),input_shape=(105,4)))

#Normalize the data
model.add(tf.keras.layers.BatchNormalization())

#Add Dense Layer which provides 10 Outputs after applying softmax
model.add(tf.keras.layers.Dense(10,input_shape = (4,), activation='relu'))
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(3, activation='softmax'))


#Comile the model
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

### Fitting the model and predicting 

```python
model.fit(X_train.values, y_train.values, epochs=100)
```

```python
model.fit(X_train.values, y_train.values, 
          validation_data=(X_test.values,y_test.values), 
          epochs=30,
          batch_size=X_train.shape[0])
```

### Report Accuracy of the predicted values

```python
model.evaluate(X_test.values,y_test.values,batch_size=X_train.shape[0])
```
